```python
# Set Pandas options to display more columns
pd.options.display.max_columns=50
train=pd.read_csv('data/1-2 가공 데이터/table_202001.csv', encoding='utf-8')     # 다 넣고
# test1=pd.read_csv('data/1-2 가공 데이터/table_202007.csv', encoding='utf-8')      # 202107 수정
# test2=pd.read_csv('data/1-2 가공 데이터/table_202008.csv', encoding='utf-8')      # 202108 수정
submission=pd.read_csv('data/1-2 가공 데이터/1-2_검증데이터셋.csv', encoding='utf-8')
train
```

### Train and Evaluate a Neural Network

```python
X=train[['Band1','Band2','Band3','Band4', 'insitu-TED0.05', 'insitu-HM', 'insitu-TD', 'insitu-TG']]
len(X.columns)
```

```python
##### Step 1 - Select data for modeling
X=train[['Band1','Band2','Band3','Band4', 'insitu-TED0.05', 'insitu-HM', 'insitu-TD', 'insitu-TG']]
y=train[['isitu-LST', 'insitu-TA']].values


##### Step 2 - Create training and testing samples
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)     # 자동이 아닌 수동으로 기간 나눠서  # 21.6까지 train, 21.9부터 val  # 새로받은


##### Step 3 - Specify the structure of a neural network
model2 = Sequential(name="Model-with-Two-Inputs")
model2.add(Input(shape=(len(X.columns),), name='Input-Layer'))
model2.add(Dense(len(X.columns), activation='relu', name='Hidden-Layer'))   # 늘리는 거 2~3
model2.add(Dense(2, activation='relu', name='Output-Layer'))


##### Step 4 - Compile the keras model
model2.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mse'],
              loss_weights=None,
              weighted_metrics=None,
              run_eagerly=None, 
              steps_per_execution=None 
             )


##### Step 5 - Fit keras model on the dataset
model2.fit(X_train.to_numpy(),       # input data
          y_train,                 # target data
          batch_size=10,      # 더 늘릴 것 백단위로
          epochs=10,         # 백이상
          verbose='auto', 
          callbacks=None, 
          validation_split=0.2, 
          #validation_data=(X_test, y_test)
          shuffle=True, 
          sample_weight=None, 
          initial_epoch=0,  
          steps_per_epoch=None, 
          validation_steps=None,
          validation_batch_size=None, 
          validation_freq=3,   #1 
          max_queue_size=10, 
          workers=1, 
          use_multiprocessing=False,
         )
```
