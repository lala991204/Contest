```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('220224_최종코드틀_TCN_50_5_0.3_256_1등파일1.csv')
```

```python
X = ['2022-03-04', '2022-03-05', '2022-03-06', '2022-03-07', '2022-03-08', '2022-03-09', '2022-03-10', \
    '2022-03-11', '2022-03-12', '2022-03-13', '2022-03-18', '2022-03-19', '2022-03-20']

y1 = df['heat_supply_day1'].values
y2 = df['heat_supply_day2'].values

plt.plot(X, y1, color='red', label='heat_supply_day1')
plt.plot(X, y2, color='blue', label='heat_supply_day2')


plt.xticks(X,fontsize=13, rotation=45)
plt.yticks(fontsize=13)
plt.legend()
plt.tight_layout()
plt.savefig('heat.png')
```
