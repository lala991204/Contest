```python
plt.scatter(df['lon'], df['lat'], color='orange', s=100)
# plt.scatter(df2['lon'], df2['lat'], color='blue', label='Group2', s=100)
# plt.scatter(df3['lon'], df3['lat'], color='green', label='Group3', s=100)

for index, stn, lon, lat in df.itertuples():
    plt.annotate("{}".format(stn),(lon, lat), size=15, xytext=(lon,lat))
# plt.hlines(y=34, xmin=126, xmax=131, linestyle='dashed', color='red')
# plt.hlines(y=37, xmin=126, xmax=131, linestyle='dashed', color='red')
plt.xlabel('lon', size=15)
plt.ylabel('lat', size=15)

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
# plt.savefig('lat.png')
plt.tight_layout()
plt.show()
```
