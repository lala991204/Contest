```python
import pandas as pd
import numpy as np
import os
```

```python
path=[i for i in os.listdir('./uv_윤정2/') if 'train' in i]
path
```

```python
data = pd.read_csv('./uv_윤정2/'+path[0])
data
```

```python
data.columns
```

```python
df=data[['stn', 'lon', 'lat']]
df
```

```python
df = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
stn1 = [13]
stn2 = [131,132,133,138,143,146,152,156,159,165]
stn3 = [105,108,112,115]

df1 = pd.DataFrame()
for i in range(len(stn1)):
    d = df[df['stn']==stn1[i]]
    df1 = pd.concat([df1, d])
    
df2 = pd.DataFrame()
for i in range(len(stn2)):
    d = df[df['stn']==stn2[i]]
    df2 = pd.concat([df2, d])
    
df3 = pd.DataFrame()
for i in range(len(stn3)):
    d = df[df['stn']==stn3[i]]
    df3 = pd.concat([df3, d])
```

```python
plt.scatter(df1['lon'], df1['lat'], color='orange', label='Group1', s=100)
plt.scatter(df2['lon'], df2['lat'], color='blue', label='Group2', s=100)
plt.scatter(df3['lon'], df3['lat'], color='green', label='Group3', s=100)

# for index, stn, lon, lat in df.itertuples():
#     plt.annotate("{}".format(stn),(lon, lat), size=15, xytext=(lon,lat))
plt.hlines(y=34, xmin=126, xmax=131, linestyle='dashed', color='red')
plt.hlines(y=37, xmin=126, xmax=131, linestyle='dashed', color='red')
plt.xlabel('lon', size=15)
plt.ylabel('lat', size=15)

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.savefig('lat.png')
plt.tight_layout()
plt.show()
```

```python
group3 = all_df[(all_df['lat'] > 37)]
group2 = all_df[(all_df['lat'] > 34) & (all_df['lat'] <= 37)]
group1 = all_df[(all_df['lat'] <= 34)]
```

```python
group1.to_csv('group1.csv', index=False)
group2.to_csv('group2.csv', index=False)
group3.to_csv('group3.csv', index=False)
```
